```python
import sys
import os
import numpy as np

# Add the src folder relative to the notebook
sys.path.append(os.path.abspath('../src'))

from metropolis_hastings import MetropolisHasting

def example(x):
    return x[0]**6*(1-x[0])**4
m = MetropolisHasting(example,[0,1])

samples = m.sampler([0],100000,20000)
print("mean")
print(np.mean(samples))
print("variance")
print(np.var(samples))


```
